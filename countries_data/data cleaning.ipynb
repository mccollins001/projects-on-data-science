# Data cleaning with pandas

## Import necessary libraries

```python
import pandas as pd
import numpy as np
```

## Read data from CSV file

```python
world = pd.read_csv('countries.csv',sep = ';')
world = world.iloc[1:,:]
```

## First and last five Rows

```python
world.head()
```

```python
world.tail()
```

## Total number of rows and columns

```python
world.shape
```

## checking for null values

```python
world.isnull().head()
```

## sum of null values for first 5 columns

```python
world.isnull().sum()
```

## Total number of null values

```python
world.isnull().sum().sum()
```

## data cleaning by back-filling the null data

```python
world = world.fillna(method = 'bfill')
```

## Filling null values by padding

```python
world = world.fillna(method = 'pad')
```

## checking for null values

```python
world.isnull().sum().sum()
```
